```python
#Importamos una de las anotaciones
import pandas as pd
file = pd.read_csv(r"vsInsectae.csv",
                   sep =";",skiprows=0)
```

```python
file.head()
```

```python
#Añadimos a un diccionario el identificador como clave y la anotación y el e valor como valor
dic={}
dummy=["cluster0"]
count=0
qseqid=[]
for i,j,z in zip(file["qseqid"],file["sseqid"],file["evalue"]) :
    #Filtramos las anotaciones que tengan un e-valor superior a cierto valor
    if "e" in str(z) or str(z) == "0.0" or str(z) == "0.001": 
        qseqid.append(i)
        dummy.append(i)
        count=count +1
        if dummy[count] != dummy[count-1]:
            dic[i]=[j,z]
```

```python
#Realizamos lo mismo con la otra anotación
import pandas as pd
columns=["Query_accesion","Target_accesion","Sequence Identity","Length","Mismatches","Gap_openings",
         "Query_start","Query_end","Target_start","Target_end","E-value","Bit_score"]
file2 = pd.read_csv(r"matchesNR.tsv",
                   sep ="\t",skiprows=0)
file2.columns=columns
file2.head()
```

```python
dic2={}
dummy2=["cluster0"]
count2=0
for i,j,z in zip(file2["Query_accesion"],file2["Target_accesion"],file2["E-value"]) :
    dummy2.append(i)
    count2=count2 +1
    if dummy2[count2] != dummy2[count2-1]:
        dic2[i]=[j,z]
```

```python
#En una lista añadimos todos los accesion que vamos a comparar
Names1=file2["Query_accesion"].unique()
Names2=file["qseqid"].unique()
Names=[]
for i,j in zip(Names1,Names2):
    if i not in Names:
        Names.append(i)
    if j not in Names:
        Names.append(j)
```

```python
#Guardamos en un archivo la comprobación con un resumen final
with open("Comparation4vnr.txt","w") as f:
    MF=0
    TF=0
    problem=[]
    for i in Names:
        f.write("Comparing annotation for:")
        f.write(i)
        f.write("\n")
        f.write("My annotation:")
        try:
            f.write(str(dic2[i][0]))
        except KeyError:
               f.write("Not found")
               problem.append(i)
               MF=MF+1
        f.write("\n")
        f.write("Their annotation:")
        try:
            f.write(str(dic[i][0]))
        except KeyError:
               f.write("Not found")
               problem.append(i)
               TF=TF+1
        f.write("\n")
        f.write("E-values")
        f.write("\n")
        f.write("Our:")
        try:
            f.write(str(dic2[i][1]))
        except KeyError:
            f.write("Not found")
        f.write("\n")
        f.write("Their:")
        try:
            f.write(str(dic[i][1]))
        except KeyError:
            f.write("Not found")
        f.write("\n")
    f.write("Clusters not found in our annotation")
    f.write(str(MF))
    f.write("\n")
    f.write("Clusters not found in their annotation")
    f.write(str(TF))
```

```python
#Para ver el resultado de cada anotación
for i in problem:
    print(i)
    try:
        print(dic2[i])
    except:
        print("no la tenemos",i)
    try:
        print(dic[i])
    except:
        print("no la tienen", i)
    print("\n")
```
