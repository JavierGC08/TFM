```python
#Se importan los paquetes necesarios

import pandas as pd
```

```python
#Se abre una de las anotaciones

file = pd.read_csv(r"vsInsectae.csv",
                   sep =";",skiprows=0)
```

```python
file.head()
```

```python
#Se añade a un diccionario el identificador como clave y la anotación y el e valor como valor

dic={}
dummy=["cluster0"]
count=0
qseqid=[]

for i,j,z in zip(file["qseqid"],file["sseqid"],file["evalue"]) :
    
    #Se filtran las anotaciones que tengan un e-valor superior a cierto valor
    if "e" in str(z) or str(z) == "0.0" or str(z) == "0.001":       
        qseqid.append(i)
        dummy.append(i)
        count=count +1
        if dummy[count] != dummy[count-1]:
            
            dic[i]=[j,z]
```

```python
#Se realiza lo mismo con la otra anotación

columns=["Query_accesion","Target_accesion","Sequence Identity","Length","Mismatches","Gap_openings",
         "Query_start","Query_end","Target_start","Target_end","E-value","Bit_score"]
file2 = pd.read_csv(r"matchesNRU.tsv",
                   sep ="\t",skiprows=0)
file2.columns=columns
file2.head()
```

```python
#Se recupera la primera anotación de cada cluster

dic2={}
dummy2=["cluster0"]
count2=0

for i,j,z in zip(file2["Query_accesion"],file2["Target_accesion"],file2["E-value"]) :
    dummy2.append(i)
    count2=count2 +1
    if dummy2[count2] != dummy2[count2-1]:
        dic2[i]=[j,z]
```

```python
#Para comprobar cuantos transcritos hay anotados

count=0
for i in dic2.keys():
    count=count+1
```

```python
print(count)
```

```python
#Para cargar otra anotación con el mismo formato

columns=["Query_accesion","Target_accesion","Sequence Identity","Length","Mismatches","Gap_openings",
         "Query_start","Query_end","Target_start","Target_end","E-value","Bit_score"]
file3 = pd.read_csv(r"matchesNR.tsv",
                   sep ="\t",skiprows=0)
file3.columns=columns
file3.head()
```

```python
# Se recupera la primera anotación de cada cluster

dic3={}
dummy3=["cluster0"]
count3=0

for i,j,z in zip(file3["Query_accesion"],file3["Target_accesion"],file3["E-value"]) :
    dummy3.append(i)
    count3=count3 +1
    if dummy3[count3] != dummy3[count3-1]:
        dic3[i]=[j,z]
```

```python
count=0
for i in dic3.keys():
    count=count+1
```

```python
print(count)
```

```python
#En una lista se añaden todos los accesion que se quiere comparar

Names1=file2["Query_accesion"].unique()
Names2=file3["Query_accesion"].unique() #si es con la otra anotación file["qseqid"].unique()
Names=[]

for i,j in zip(Names1,Names2):
    if i not in Names:
        Names.append(i)
    if j not in Names:
        Names.append(j)
```

```python
'''
Se guarda en un archivo la comprobación con un resumen final, además de los e valores que son distintos
por si se quiere comparar dos anotaciones muy parecidas
'''

with open("Comparation5nrvnru.txt","w") as f:
    
    MF=0
    TF=0
    problem=[]
    EV=[]
    EV2=[]
    cluster=[]
    cluster2=[]
    
    for i in Names:
        
        f.write("Comparing annotation for:")
        f.write(i)
        f.write("\n")
        f.write("My annotation:")
        
        '''
        Se utiliza un sistema de excepciones porque si una anotación no se encuentra en uno de los dos archivos da error.
        En vez de eso se quiere guardar como no encontrado o no anotado
        '''
        
        try:
            f.write(str(dic2[i][0]))
        except KeyError:
               f.write("Not found")
               problem.append(i)
               MF=MF+1
        f.write("\n")
        f.write("Their annotation:")
        try:
            f.write(str(dic3[i][0]))
        except KeyError:
               f.write("Not found")
               problem.append(i)
               TF=TF+1
        f.write("\n")
        f.write("E-values")
        f.write("\n")
        f.write("Our:")
        try:
            f.write(str(dic2[i][1]))
            EV.append(str(dic2[i][1]))
        except KeyError:
            f.write("Not found")
            EV.append("none")
        f.write("\n")
        f.write("Their:")
        try:
            f.write(str(dic3[i][1]))
            EV2.append(str(dic3[i][1]))
        except KeyError:
            f.write("Not found")
            EV2.append("none")
        f.write("\n")
        
    f.write("Clusters not found in our annotation")
    f.write(str(MF))
    f.write("\n")
    f.write("Clusters not found in their annotation")
    f.write(str(TF))
```

```python
#Para ver el resultado de cada anotación

for i in problem:
    print(i)
    try:
        print(dic2[i])
    except:
        print("no la tenemos",i)
    try:
        print(dic3[i])
    except:
        print("no la tienen", i)
    print("\n")
```

```python
#Comparar aquellos clusters que tienen e valores distintos, solo sirve para anotaciones muy parecidas

for i in Names:
    try:
        if dic2[i] != dic3[i]:
            print(i,dic2[i],dic3[i])
    except:
        pass
```
