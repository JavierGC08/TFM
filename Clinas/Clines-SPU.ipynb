**Script para filtrar el archivo de clinas por aquellas que fuesen significativas en al menos una variable y que además estuviese anotada.**

```python
#Importamos el archivo original del análisis de clinas
import pandas as pd
filecsv = pd.read_csv(r"Clines.csv",
                   sep =",",skiprows=0)
```

```python
filecsv.head()
```

```python
CTF=[]
CFT=[]
CTT=[]
PTF=[]
PFT=[]
PTT=[]
for a,b,c,d,e in zip(filecsv["Portalet_centre_sig_clineplots_minus20"],filecsv["PaysBasco_centre_sig_clineplots_minus20"],
                     filecsv["Portalet_v_sig_clineplots_minus20"],filecsv["PaysBasco_v_sig_clineplots_minus20"],filecsv["Portalet_locus"]):
    a=str(a)
    b=str(b)
    c=str(c)
    d=str(d)
    e=str(e)
    #print(a,b,c,d,e)
    if a =="True" and b == "True":
        CTT.append(e[e.find("_")+1:])
    if a == "True" and b=="False":
        CTF.append(e[e.find("_")+1:])
    if a=="False" and b=="True":
        CFT.append(e[e.find("_")+1:])
    if c=="True" and d == "True":
        PTT.append(e[e.find("_")+1:])
    if c=="True" and d=="False":
        PTF.append(e[e.find("_")+1:])
    if c=="False" and d=="True":
        PFT.append(e[e.find("_")+1:])
```

```python
with open ("grasshopperRef.positions", "r") as f:
    fileG=f.readlines()
    ref={}
    for i in fileG:
        if "chr2" in i:
            break
        pos=i.find(":")+1
        end=i.find("\t")
        ID=i[pos:end]
        ref[ID]=[i[(end+1):-1]]
L=[]
with open ("snp_names","r") as f2:
    file2=f2.readlines()
    for i in file2:
        L.append(i)
with open ("snp_names","r") as f2:
    file2=f2.readlines()
    count=0
    for i in ref.keys():
        start=[]
        start=str(ref[i])
        pos=start.find(":")
        pos2=start.find("-")
        pri=int(start[pos+1:pos2])
        sec=int(start[pos2+1:-2])
        number=L[count][:(L[count].find("\t"))]
        if int(number) in range(pri,sec):
            ref[i].append(number)
            count=count+1
            if count == (len(L)+1):
                break
```

```python
for i in ref.keys():
    try:
        if ref[i][1] in CTT:
            ref[i].append("CTT")
        if ref[i][1] in CTF:
            ref[i].append("CTF")
        if ref[i][1] in CFT:
            ref[i].append("CFT")
        if ref[i][1] in PTT:
            ref[i].append("PTT")
        if ref[i][1] in PTF:
            ref[i].append("PTF")
        if ref[i][1] in PFT:
            ref[i].append("PFT")
    except:
        pass
```

```python
columns=["Query_accesion","Target_accesion","Sequence Identity","Length","Mismatches","Gap_openings",
         "Query_start","Query_end","Target_start","Target_end","E-value","Bit_score"]
```

```python
fileGN= pd.read_csv(r"matchesSPU.tsv",
                   sep ="\t",skiprows=0,header=None)
```

```python
fileGN.columns=columns
```

```python
fileGN.head()
```

```python
dic2={}
dummy2=["cluster0"]
count2=0
for i,j,z in zip(fileGN["Query_accesion"],fileGN["Target_accesion"],fileGN["E-value"]) :
    dummy2.append(i)
    count2=count2 +1
    if dummy2[count2] != dummy2[count2-1]:
        dic2[i]=[j,z]
```

```python
Names=fileGN["Query_accesion"].unique()
```

```python
header=("SNP","CLUSTER","CENTER","SLOPE","ANNOTATION")
df=pd.DataFrame(columns=header)
```

```python
SNPL=[]
clusterL=[]
CenterL=[]
SlopeL=[]
AnnotationL=[]
for i in Names:
        SNP=""
        prueba=""
        cluster=""
        A1=""
        center=""
        pen=""
        try:
            prueba=ref[i][2]
            SNP=ref[i][1]
            cluster=i
            try:
                A1=dic2[i][0]
            except:
                A1="Null"
            try:
                if "C" in ref[i][2]:
                    center=ref[i][2]
                else:
                    pen=ref[i][2]
                try:
                    if "C" in ref[i][3]:
                        center=ref[i][3]
                    else:
                        pen=ref[i][3]
                except:
                    if "C" not in center:
                        center="Null"
                    elif "P" not in pen:
                        pen="Null"
            except:
                center="Null"
                pen="Null"
            #print(i,SNP,cluster,center,pen)
            SNPL.append(SNP)
            clusterL.append(cluster)
            CenterL.append(center)
            SlopeL.append(pen)
            AnnotationL.append(A1)
        except:
            pass
```

```python
df["SNP"]=SNPL
df["CLUSTER"]=clusterL
df["CENTER"]=CenterL
df["SLOPE"]=SlopeL
df["ANNOTATION"]=AnnotationL
```

```python
df.head()
```

```python
for i in df["SNP"]:
    name=str("chr1_"+i)
    posicion1=df.loc[df["SNP"] == i].index[0]
    posicion2=filecsv.loc[filecsv["Portalet_locus"] == name].index[0]
    c=df.iloc[posicion1]["CENTER"]
    c2=filecsv.iloc[posicion2]["Portalet_centre_sig_clineplots_minus20"]
    c3=filecsv.iloc[posicion2]["PaysBasco_centre_sig_clineplots_minus20"]
    p=df.iloc[posicion1]["SLOPE"]
    p2=filecsv.iloc[posicion2]["Portalet_v_sig_clineplots_minus20"]
    p3=filecsv.iloc[posicion2]["PaysBasco_v_sig_clineplots_minus20"]
    if "Null" == c:
        if str(c2) != "False":
            print("error",i,c,c2)
            break
        if str(c3) != "False":
            print("error",i,c,c3)
            break
    if "C" in str(c):
        if str(c)[1] != str(c2)[0]:
            print("error",i,c,c2)
        if str(c)[2] != str(c3)[0]:
            print("error",i,c,c3)
    if "Null" == p:
        if str(p2) != "False":
            print("error",i,p,p2)
            break
        if str(p3) != "False":
            print("error",i,p,p3)
            break
    if "P" in str(p):
        if str(p)[1] != str(p2)[0]:
            print("error",i,p,p2)
            break
        if str(p)[2] != str(p3)[0]:
            print("error",i,p,p3)
            break
print("Center and Slope are correct")
```
