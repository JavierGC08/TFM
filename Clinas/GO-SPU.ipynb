**Script para obtener la lista de genes total  y la de cada una de los grupos de interés, además de los backgrounds**

```python
#Importamos el archivo de las clinas significativas, cuyos identificadores serán los que introducimos en el análisis
#funcional

import pandas as pd
filecsv = pd.read_csv(r"ClineResultsSPU",
                   sep ="\t",skiprows=0)
```

```python
filecsv.head()
```

```python
with open ("genelist","w") as f:
    for i in filecsv["ANNOTATION"]:
        f.write(i[:(i.find("."))])
        f.write("\n")
```

```python
#Iteramos sobre el dataframe de las clinas y vamos guardando en diferentes listas los identificadores correspondientes

LCPP=[]
CPTT=[]
LCTPT=[]
CCFPF=[]
CTPT=[]
CT=[]
PT=[]
C_TP_T=[]
C_T=[]
P_T=[]
for i,a,b in zip(filecsv["ANNOTATION"],filecsv["CENTER"],filecsv["SLOPE"]):
    
    #La herramienta de DAVID tiene algunos problemas con el siguiente identificador, con lo que lo eliminamos, aún teniéndolo en cuenta
    
    if str(i) != "P02299.4":
        if a == "CTT":
            LCPP.append(i)
        if b == "PTT":
            CPTT.append(i)
        if a =="CTF" and b == "PTF":
            LCTPT.append(i)
        if a =="CFT" and b == "PFT":
            CCFPF.append(i)
        if a =="CTF" or a=="CTT":
            CT.append(i)
        if b=="PTF" or b=="PTT":
            PT.append(i)
        if a =="CTF" or a=="CTT":
            if b=="PTF" or b=="PTT":
                CTPT.append(i)
        if a =="CFT" or a=="CTT":
            C_T.append(i)
        if b=="PFT" or b=="PTT":
            P_T.append(i)
        if a =="CFT" or a=="CTT":
            if b=="PFT" or b=="PTT":
                C_TP_T.append(i)
```

```python
#Eliminamos los espacios en blanco que tienen algunas listas.

while "" in LCPP:
    LCPP.remove("")
while "" in CPTT:
    CPTT.remove("")
while "" in LCTPT:
    LCTPT.remove("")
while "" in CCFPF:
    CCFPF.remove("")
while "" in CT:
    CT.remove("")
while "" in PT:
    PT.remove("")
while "" in CTPT:
    CTPT.remove("")
while "" in C_T:
    C_T.remove("")
while "" in P_T:
    P_T.remove("")
while "" in C_TP_T:
    C_TP_T.remove("")
```

```python
#Creamos dos listas más, una para cada transecto, para realizar análisis adicionales

PV=[]
for i in C_T:
    PV.append(i)
for i in P_T:
    if i not in PV:
        PV.append(i)
for i in C_TP_T:
    if i not in PV:
        PV.append(i)
```

```python
Por=[]
for i in CT:
    Por.append(i)
for i in PT:
    if i not in Por:
        Por.append(i)
for i in CTPT:
    if i not in Por:
        Por.append(i)
```

```python
#Comprobamos la longitud de las listas

print(len(LCPP),len(CPTT),len(LCTPT),len(CCFPF),len(CT),len(PT),len(CTPT),len(C_T),len(P_T),len(C_TP_T))
```

```python
#Volvemos a realizar la relación entre cluster y SNP

with open ("grasshopperRef.positions", "r") as f:
    fileG=f.readlines()
    ref={}
    for i in fileG:
        if "chr2" in i:
            break
        pos=i.find(":")+1
        end=i.find("\t")
        ID=i[pos:end]
        ref[ID]=[i[(end+1):-1]]
L=[]
with open ("snp_names","r") as f2:
    file2=f2.readlines()
    for i in file2:
        L.append(i)
with open ("snp_names","r") as f2:
    file2=f2.readlines()
    count=0
    for i in ref.keys():
        start=[]
        start=str(ref[i])
        pos=start.find(":")
        pos2=start.find("-")
        pri=int(start[pos+1:pos2])
        sec=int(start[pos2+1:-2])
        number=L[count][:(L[count].find("\t"))]
        if int(number) in range(pri,sec):
            ref[i].append(number)
            count=count+1
            if count == (len(L)+1):
                break
```

```python
#Abrimos el archivo de nuestra anotación con swissprot, que servirá como background general

fileIns = pd.read_csv(r"matchesSPU.tsv",
                   sep ="\t",skiprows=0,header=None)
fileIns.head()
```

```python
#Nos quedamos solo con el primer hit por cluster
dummy=["cluster0"]
count2=0
for i,j,z in zip(fileIns[0],fileIns[1],fileIns[10]) :
    dummy.append(i)
    count2=count2 +1
    if dummy[count2] != dummy[count2-1]:
        ref[i].append(j)
```

```python
#Guardamos en un archivo los identificadores
with open ("backgroundWholeSPU","w") as f:
    for i in ref.keys():
        try:
            name=ref[i][2]
            f.write(name[:(name.find("."))])
            f.write("\n")
        except:
            pass
```
