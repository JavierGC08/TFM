```python
#Se importan los paquetes
import pandas as pd
```

```python
#Se definen las funciones que se van a usar
def relation():
    with open ("grasshopperRef.positions", "r") as f:
        fileG=f.readlines()
        ref={}
        for i in fileG:
            if "chr2" in i:
                break
            pos=i.find(":")+1
            end=i.find("\t")
            ID=i[pos:end]
            ref[ID]=[i[(end+1):-1]]
    L=[]
    with open ("snp_names","r") as f2:
        file2=f2.readlines()
        for i in file2:
            L.append(i)
    with open ("snp_names","r") as f2:
        file2=f2.readlines()
        count=0
        for i in ref.keys():
            start=[]
            start=str(ref[i])
            pos=start.find(":")
            pos2=start.find("-")
            pri=int(start[pos+1:pos2])
            sec=int(start[pos2+1:-2])
            number=L[count][:(L[count].find("\t"))]
            if int(number) in range(pri,sec):
                ref[i].append(number)
                count=count+1
                if count == (len(L)+1):
                    break
    return ref
```

```python
'''
Se importan el archivo de las clinas significativas, cuyos identificadores serán los que se introduzcan
en el análisis funcional
'''
filecsv = pd.read_csv(r"ClineResultsNR",
                   sep ="\t",skiprows=0)
```

```python
with open ("genelist2","w") as f:
    for i in filecsv["ANNOTATION"]:
        f.write(i[:(i.find("."))])
        f.write("\n")
```

```python
#Se itera sobre el dataframe de las clinas y vamos guardando en diferentes listas los identificadores correspondientes

CTT=[]
PTT=[]
CTFPTF=[]
CFTPFT=[]
CTPT=[]
CT=[]
PT=[]
C_TP_T=[]
C_T=[]
P_T=[]
for i,a,b in zip(filecsv["ANNOTATION"],filecsv["CENTER"],filecsv["SLOPE"]):
    
    #La herramienta de DAVID tiene algunos problemas con el siguiente identificador, 
    #con lo que lo se elimina, aún teniéndolo en cuenta
    
    if str(i) != "P02299.4":
        if a == "CTT":
            CTT.append(i)
        if b == "PTT":
            PTT.append(i)
        if a =="CTF" and b == "PTF":
            CTFPTF.append(i)
        if a =="CFT" and b == "PFT":
            CFTPFT.append(i)
        if a =="CTF" or a=="CTT":
            CT.append(i)
        if b=="PTF" or b=="PTT":
            PT.append(i)
        if a =="CTF" or a=="CTT":
            if b=="PTF" or b=="PTT":
                CTPT.append(i)
        if a =="CFT" or a=="CTT":
            C_T.append(i)
        if b=="PFT" or b=="PTT":
            P_T.append(i)
        if a =="CFT" or a=="CTT":
            if b=="PFT" or b=="PTT":
                C_TP_T.append(i)
```

```python
#Se eliminan los espacios en blanco que tienen algunas listas.

while "" in CTT:
    CTT.remove("")
while "" in PTT:
    PTT.remove("")
while "" in CTFPTF:
    CTFPTF.remove("")
while "" in CFTPFT:
    CFTPFT.remove("")
while "" in CT:
    CT.remove("")
while "" in PT:
    PT.remove("")
while "" in CTPT:
    CTPT.remove("")
while "" in C_T:
    C_T.remove("")
while "" in P_T:
    P_T.remove("")
while "" in C_TP_T:
    C_TP_T.remove("")
```

```python
#Se crean dos listas más, una para cada transecto, para realizar análisis adicionales

PV=[]
for i in C_T:
    PV.append(i)
for i in P_T:
    if i not in PV:
        PV.append(i)
for i in C_TP_T:
    if i not in PV:
        PV.append(i)
```

```python
Por=[]
for i in CT:
    Por.append(i)
for i in PT:
    if i not in Por:
        Por.append(i)
for i in CTPT:
    if i not in Por:
        Por.append(i)
```

```python
#Se abre el archivo de nuestra anotación con swissprot, que servirá como background general

fileIns = pd.read_csv(r"matchesSPU.tsv",
                   sep ="\t",skiprows=0,header=None)
fileIns.head()
```

```python
#Se guarda solo con el primer hit por cluster
ref=relation()
dummy=["cluster0"]
count2=0
for i,j,z in zip(fileIns[0],fileIns[1],fileIns[10]) :
    dummy.append(i)
    count2=count2 +1
    if dummy[count2] != dummy[count2-1]:
        ref[i].append(j)
```

```python
#Se guarda en un archivo los identificadores
with open ("backgroundWholeSPU","w") as f:
    for i in ref.keys():
        try:
            name=ref[i][2]
            f.write(name[:(name.find("."))])
            f.write("\n")
        except:
            pass
```

```python
#Hay que quedarse solo con las clinas que estén anotadas
fileIns = pd.read_csv(r"matchesSPU.tsv",
                   sep ="\t",skiprows=0,header=None)

dummy=["cluster0"]
count2=0
```

```python
#Se guardan las clinas que esten anotadas y se vuelve a generar el diccionario al haberse modificado
ref=relation()
for i in filecsv["Portalet_locus"]:
    locus=i[(i.find("_"))+1:]
    for j in ref.keys():
        try:
            if ref[j][1]==locus:
                for i,j,z in zip(fileIns[0],fileIns[1],fileIns[10]) :
                    dummy.append(i)
                    count2=count2 +1
                    if dummy[count2] != dummy[count2-1]:
                        ref[i].append(j)
        except:
            pass
```

```python
#Se guarda la anotación de las clinas
with open ("backgroundClineSPU","w") as f:
    for i in ref.keys():
        print(ref[i])
        try:
            name=ref[i][2]
            f.write(name[:(name.find("."))])
            f.write("\n")
        except:
            pass
```
