```python
#Importamos los paquetes que vamos a usar
import pandas as pd
```

```python
# A partir de un csv con los posibles genes lo guardamos en un dataframe de pandas
file = pd.read_csv(r"Posible genes.csv",
                   sep =";",skiprows=0,)
file.head()
```

```python
#Creamos un nuevo dataframe en el que vamos a guardar las columnas que nos interesan
columns = ["Nombre del gen","Grupo"]
df = pd.DataFrame(columns=columns)
df.columns=columns
```

```python
#Utilizamos un bucle for con iteracion sobre dos listas a la vez, de esta forma, podemos guardar en orden el grupo al que pertenece el gen.
Nombre=[]
Grupo=[]
for i,j in zip(file["Nombre del gen"],file["Grupo"]):
    #Algunos genes tienen varios nombres, como no sabemos cual tiene en la base de datos sobre la que buscamos, nos quedamos con todos los posibles.
    if "," in i  :
        dummy=i.split(",") 
        ListaDef=[]
        for i in dummy:
            ListaDef.append(i.lstrip())
        for i in ListaDef:
            Nombre.append(i)
            Grupo.append(j)
    else:
        Nombre.append(i)
        Grupo.append(j)
#Guardamos en el dataframe las dos listas
df["Nombre del gen"]=Nombre
df["Grupo"]=Grupo
```

```python
#Creamos un segundo dataframe que tenga en una única columna la entrada que necesita GNfish para buscar

bus=[]
for i,j in zip(df["Nombre del gen"], df["Grupo"]):
    
    #Como vamos a buscar solo en Insect, en caso de que no tuviese grupo en el paper añadimos siempre Insect.
    if str(j) == "nan":
        print(j)
        name=str(i)+" " +"("+ "AND" +" "+"Insect"+ "[Organism]" +")"
    else:
        name=str(i)+" " +"("+ "AND" +" "+str(j)+ "[Organism]" +")"
    bus.append(name)
columns = ["Busqueda"]
df2 = pd.DataFrame(columns=columns)
df2.columns=columns
df2["Busqueda"]=bus
```
